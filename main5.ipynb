```python
import pandas as pd
from konlpy.tag import Komoran
from tqdm.notebook import tqdm
import pickle
from collections import Counter

kom = Komoran()

def counter_dump_between(start, end):
    for year in tqdm(range(start, end+1), position=0):
        df = pd.read_parquet('./data/chosun_1980.parquet')
        data = df.loc[:, 'body_archaic_hangul']

        r = Counter()
        for text in tqdm(data, position=1):
            try:
                morphs = kom.morphs(text)
                r += Counter(morphs)
            except Exception as error:
                print(year)
                print(error)
                print(text)

        with open(f'./data/komoran_counter_{year}.pickle', 'wb') as f:
            pickle.dump(r, f)
```
