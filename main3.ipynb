```python
import pickle
from tqdm.notebook import tqdm
```

```python
datasets = ['chosun_first', 'chosun_second', 'chosun_last', 'donga_first', 'donga_second', 'donga_last']
```

```python
from collections import Counter

r = {}
for dataset in tqdm(datasets):
    with open('./data/' + dataset + '.pkl', 'rb') as f:
        r[dataset] = Counter(pickle.load(f))

with open('./data/counter.pkl', 'wb') as f:
    pickle.dump(r, f)
```

```python
with open('./data/counter.pkl', 'rb') as f:
    data0 = Counter(pickle.load(f))
```

```python
R = {}
for dataset in tqdm(r.keys()):
    for k,v in r[dataset].items():
        if k[0] in R.keys():
            R[k[0]].append((dataset, v))
        else:
            R[k[0]] = [(dataset, v)]

with open('./data/counter1.pkl', 'wb') as f:
    pickle.dump(R, f)
```

```python
with open('./data/counter1.pkl', 'rb') as f:
    R = Counter(pickle.load(f))
```

```python
_R = {}
for word, v in R.items():
    if set(datasets) == set([x[0] for x in v]):
        r1 = []
        for dataset in datasets[:3]:
            v1 = filter(lambda x: x[0] == dataset, v)
            s = sum([x[1] for x in v1])
            r1.append(s)
        r2 = []
        for dataset in datasets[3:]:
            v1 = filter(lambda x: x[0] == dataset, v)
            s = sum([x[1] for x in v1])
            r2.append(s)
        s = sum(r1 + r2)
        r = {'chosun': r1, 'donga': r2, 'sum': s}
        _R[word] = r
```
